```python
import pandas as pd
import numpy as np
```

```python
mta = pd.read_csv('~/ds4a_workspace/final 9/mta_trips.csv.gz', nrows=100)
uber = pd.read_csv('~/ds4a_workspace/final 9/uber_trips_2014.csv.gz', nrows=100)
green = pd.read_csv('~/ds4a_workspace/final 9/green_trips.csv.gz', nrows=100)
yellow = pd.read_csv('~/ds4a_workspace/final 9/yellow_trips.csv.gz', nrows=100)
```

```python
def clean_mta(df):
    df['interval'] = pd.to_datetime(df['datetime']).dt.hour
    df['ride_type'] = 'mta'
    columns = df.columns
    repeats = df['new_entries'].values
    df = pd.DataFrame(np.repeat(df.values, repeats, axis=0))
    df.columns = columns    
    return df[['ride_type','interval','latitude','longitude']]
```

```python
mta = clean_mta(mta)
mta.head()
```

```python
def clean_uber(df):
    df['interval'] = pd.to_datetime(df['pickup_datetime']).dt.hour
    df.columns = [['datetime','latitude','longitude','base','interval']]
    df['ride_type'] = 'uber'
    return df[['ride_type','interval','latitude','longitude']]
```

```python
uber = clean_uber(uber)
uber.head()
```

```python
def clean_taxi(df, taxi_type):
    df['interval'] = pd.to_datetime(df['pickup_datetime']).dt.hour
    df = df[['interval','pickup_latitude','pickup_longitude']]
    df.columns = ['interval','latitude','longitude']
    df['ride_type'] = taxi_type
    return df[['ride_type','interval','latitude','longitude']]
```

```python
green = clean_taxi(green, 'green')
green.head()
```

```python
yellow = clean_taxi(yellow, 'yellow')
yellow.head()
```

```python
all_rides = pd.concat([mta,uber,green,yellow])
```
